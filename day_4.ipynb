```python
import pandas as pd
```

```python
cancer = pd.read_csv("./datasets/cancer/dataset.csv")

cancer.head()
```

```python
cancer.describe()
```

```python
cancer.info()
```

```python
cancer = cancer.drop(cancer.columns[32], axis=1)
cancer.info()
```

```python
cancer["diagnosis"] = cancer["diagnosis"].map({"M": 1, "B": 0})
```

```python
cancer["diagnosis"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data=cancer, x="radius_mean", hue="diagnosis", ax=axes[0], kde=True)
sns.histplot(data=cancer, x="texture_mean", hue="diagnosis", ax=axes[1], kde=True)

plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x="diagnosis", y="radius_mean", data=cancer)
plt.show()
```

```python
plt.figure(figsize=(25, 20))
correlation_matrix = cancer.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(
    cancer,
    hue="diagnosis",
    vars=["radius_mean", "texture_mean", "perimeter_mean", "area_mean"],
)

plt.show()
```

### Findings

1. The correlation between geometrical attributes associated with breasts and breast cancer is pretty evident from the data
2. Data is not scaled, so it needs to be


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = cancer.drop(columns=["diagnosis", "id"], axis=1)
y = cancer["diagnosis"]

scaled_X_features = scaler.fit_transform(X)
scaled_X = pd.DataFrame(scaled_X_features, columns=X.columns, index=X.index)
```

```python
scaled_X.describe()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, Y_test = train_test_split(
    scaled_X, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.svm import SVC

linear_svm_classifier = SVC(kernel="linear")
linear_svm_classifier.fit(X_train, y_train)
```

```python
linear_y_pred = linear_svm_classifier.predict(X_test)

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

linear_accuracy = accuracy_score(Y_test, linear_y_pred)
print(linear_accuracy)
print(classification_report(Y_test, linear_y_pred))
```

```python
rfb_svm_classifier = SVC(kernel="rbf")
rfb_svm_classifier.fit(X_train, y_train)
```

```python
rfb_y_pred = rfb_svm_classifier.predict(X_test)

rfb_accuracy = accuracy_score(Y_test, rfb_y_pred)
print(rfb_accuracy)
print(classification_report(Y_test, rfb_y_pred))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
axes[0].set_title("Linear SVC Confusion Matrix")
axes[1].set_title("RBF SVC Confusion Matrix")


ConfusionMatrixDisplay.from_estimator(
    linear_svm_classifier, X_test, Y_test, ax=axes[0], cmap="Blues"
)
ConfusionMatrixDisplay.from_estimator(
    rfb_svm_classifier, X_test, Y_test, ax=axes[1], cmap="Purples"
)

plt.tight_layout()
plt.show()
```

### Findings

1. Using different kernels leads to difference in the accuracy score
2. RBF Kernel performs better considering the data has a lot of dimensions to work with


```python
import numpy as np
from sklearn.model_selection import GridSearchCV

param_grid = {"C": [0.1, 1, 10, 100, 1000], "gamma": [1, 0.1, 0.01, 0.001, 0.0001]}


grid_search = GridSearchCV(
    estimator=rfb_svm_classifier,
    param_grid=param_grid,
    scoring="accuracy",
    cv=5,
    verbose=2,
    n_jobs=-1,
)

grid_search.fit(X_train, y_train)

print(f"The best parameters found are: {grid_search.best_params_}")
print(f"The best cross-validated accuracy is: {grid_search.best_score_:.4f}")

best_svc_preds = grid_search.predict(X_test)


print(
    f"Test set accuracy of the best model: {accuracy_score(Y_test, best_svc_preds):.4f}"
)
print("\nClassification Report on Test Set:")
print(classification_report(Y_test, best_svc_preds))
```

```python
ConfusionMatrixDisplay.from_estimator(grid_search, X_test, Y_test, cmap="Greens")
```

### Findings

1. Grid Search leads to better hyperparameter optimization, that is automatic in nature
2. It leads to more accurate model training
3. _C_ is for cost of misclassification
4. _gamma_ is for influence of each point on the decision boundary
